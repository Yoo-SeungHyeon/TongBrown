```python
import matplotlib.pyplot as plt
```

```python
x=[30,20,60,80,40,50,60,30,70,60]
y=[73,50,128,170,87,108,135,69,148,132]
```

```python
plt.plot(x,y,'ro')
```

```python
n = int(len(x))
Sx = sum(x)
Sy = sum(y)

xy=[]
for i in range(n):
    xy.append(x[i]*y[i])

Sxy = sum(xy)

xx = []
for i in range(n):
    xx.append(x[i]*x[i])

Sxx = sum(xx)
```

```python
print(n,Sx,Sy,Sxy,Sxx)
```

```python
a= Sx*Sy/n
print(a)
```

```python
b= Sx*Sx/n
print(b)
```

```python
aa = Sxy - a
bb = Sxx - b
b1 = aa/bb
```

```python
b1
```

```python
b0 = Sy/n - b1*Sx/n
```

```python
b0
```

```python
x2 = range(15,85)
y2=[]
for i in range(len(x2)):
    y2.append(b0+b1*x2[i])
```

```python
plt.plot(x,y,'ro')
plt.plot(x2,y2)
```
