```python
from nltk.tag import pos_tag
import nltk
ps = PorterStemmer()
nltk.download('averaged_perceptron_tagger')
```

```python
sentence = "Emma refused to permit us to obtain the refuse permit"
tagged_list = pos_tag(word_tokenize(sentence))
tagged_list
```

```python
Word=[]
View=[]
U_date=[]
Link=[]
E_date=[]
Par=[]
i=0

for title in titles:
    tag_list = pos_tag(word_tokenize(title))
    for tl in tag_list:
        doc=tl[0]
        par=tl[1]
#         porter = [porter_stemmer.stem(i) for i in doc]
        por = ps.stem(doc)
        Word.append(por)
        Par.append(par)
        View.append(view[i])
        U_date.append(u_date[i])
        Link.append(link[i])
        E_date.append(e_date[i])
    i = i+1
```

```python
newdic = {'WORD' : Word, 'PART' : Par, 'VIEW' : View, 'U_DATE' : U_date, 'LINK' : Link, 'E_DATE' : E_date}
```

```python
N_df = pd.DataFrame(newdic)
```

```python
essendic = {'word':Word, 'part':Par,'link':Link}
```

```python
essen_df = pd.DataFrame(essendic)
```
